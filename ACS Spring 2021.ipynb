```python
import param_search, glob

def find_work_dir(x):
    if 'work_dir' in x and not pd.isnull(x['work_dir']):
        return x['work_dir']
    f = glob.glob('*/{}.*'.format(x['job_id']))
    if f:
        return os.path.dirname(f[0])
    # else cancelled before running

def find_job_state(x):
    if 'job_state' in x and not pd.isnull(x['job_state']):
        return x['job_state']
    if x['work_dir'] is None:
        return 'CAN'
    if x['stderr'] is not None:
        return 'ERR'
    return 'OK'

def catch_exc(func, exc_type=Exception, default=None):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except exc_type as e:
            return default
    return wrapper

jobs = pd.DataFrame(dict(job_id=job_ids))
qstat = param_search.status(user='mtr22')
qstat['job_id'] = qstat['job_id'].map(pd.to_numeric)
jobs = jobs.merge(qstat, on='job_id', how='left')
jobs['work_dir'] = jobs.apply(find_work_dir, axis=1)
jobs['job_name'] = jobs['work_dir'].map(catch_exc(os.path.basename, TypeError))
jobs['job_file'] = jobs['work_dir'].map(lambda x: '{}/generate.sh'.format(x))
jobs['stdout_file'] = jobs[['work_dir', 'job_id']].apply(lambda x: '{work_dir}/{job_id}.stdout'.format(**x), axis=1)
jobs['stderr_file'] = jobs[['work_dir', 'job_id']].apply(lambda x: '{work_dir}/{job_id}.stderr'.format(**x), axis=1)
jobs['stdout'] = jobs['stdout_file'].map(catch_exc(param_search.job_output.read_stdout_file, FileNotFoundError))
jobs['stderr'] = jobs['stderr_file'].map(catch_exc(param_search.job_output.read_stderr_file, FileNotFoundError))
jobs['job_state'] = jobs.apply(find_job_state, axis=1)
jobs.set_index(['job_id', 'job_name'])[['job_state', 'stdout', 'stderr']]
```

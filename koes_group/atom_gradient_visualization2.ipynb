Atom Gradient Visualization

```python
import py3Dmol

def split_xyz(xyz, threshold):
    count1, count2 = 0, 0
    atoms1, atoms2 = [], []
    for i, line in enumerate(xyz.split('\n')):
        if i < 2: continue
        data = line.split()
        if float(data[-1])**2 + float(data[-2])**2 + float(data[-3])**2 <= threshold**2:
            count1 += 1
            atoms1.append(line)
        else:
            count2 += 1
            atoms2.append(line)
    xyz1 = '\n'.join([str(count1), ''] + atoms1)
    xyz2 = '\n'.join([str(count2), ''] + atoms2)
    return xyz1, xyz2

def view_gradient_animation(view, rec_fname, lig_fname,
                     rec_style={'stick': {'radius': 0.1}},
                     lig_style={'stick': {'colorscheme': 'greenCarbon'}},
                     threshold=1.0,
                     loop='forward'):
    with open(rec_fname, 'r') as f:
        rec_xyz = f.read()
    with open(lig_fname, 'r') as f:
        lig_xyz = f.read()
    
    view.clear()
    #view.setBackgroundColor(0)
    
    view.addModel(lig_xyz, 'xyz')
    view.setClickable({}, True)
    view.addModel(rec_xyz, 'xyz')

    view.setStyle({}, rec_style)
    view.setStyle({'clickable': True}, lig_style)
    
    view.render()

    view.vibrate(10, 1)
    view.animate({'loop': loop})
    view.zoomTo()
    return view
```

```python
view = py3Dmol.view(width=900, height=600)
```

```python
view_gradient_animation(view, 'xyz/csar_set1_102_rec.xyz', 'xyz/csar_set1_102_lig.xyz').show()
```
